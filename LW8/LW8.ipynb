**Тезис 1**
В Google Play Store большое количество раличных приложений на различные категории. Но большая часть из них приходится на игры.



```python
games = pd.read_csv('googleplaystore.csv')

```

```python
Category = games.groupby('Category').size().plot(kind='bar')
Category.set_title('Классификация приложений Google Play Store')
Category.set_ylabel('Количество приложений')
Category.set_xlabel('Категория приложения')
plt.show()
```

Тезис опровергнут. Большая часть приложений категории "Семья"
